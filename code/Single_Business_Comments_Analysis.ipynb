# Single Business Comments Analysis

```python
import pandas as pd
import json
```

```python
all_reviews = {}


with open('../data/review_city_Chinese.json', 'r') as review:
    for line in review:
        parsed = json.loads(line)
        business_id = parsed['business_id']
        parsed.pop('business_id')
        if business_id not in all_reviews:
            all_reviews[business_id] = [parsed]
        else:
            all_reviews[business_id].append(parsed)
```

```python
len(all_reviews)
```

# Focus on One Business

```python
b1_id = 'UPSzbM6D2SkarWAINq1I9A'
b1_json = json.dumps(all_reviews[b1_id],indent=4)
```

```python
len(all_reviews[b1_id])
```

## Convert to DataFrame

```python
b1_review = pd.read_json(b1_json)
b1_review.head()
```

## Split Reviews 

```python
# good
b1_review_good = b1_review[b1_review['stars'] > 3]
b1_review_good.head()
```

```python
# moderate
b1_review_moderate = b1_review[b1_review['stars'] == 3]
b1_review_moderate.head()
```

```python
# bad
b1_review_bad = b1_review[b1_review['stars'] < 3]
b1_review_bad.head()
```

## Merge Texts

```python
b1_text_good = ''
for index, row in b1_review_good.iterrows():
    b1_text_good += row['text'].lower()

print(b1_text_good)
```

```python
b1_text_moderate = ''
for index, row in b1_review_moderate.iterrows():
    b1_text_moderate += row['text'].lower()
print(b1_text_moderate)
```

```python
b1_text_bad = ''
for index, row in b1_review_bad.iterrows():
    b1_text_bad += row['text'].lower()
print(b1_text_bad)
```

## Analyse Text

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from collections import Counter
```

```python
stop_words = set(stopwords.words('english')) | set(['!', '(', ')', ',', '.', '?', '#', '$', 
                                                    '*', '@', ':', 'go', "'s", "n't", 'one', 
                                                    'ordered', 'order', "'ve", '...', 'place', 'get',
                                                   'also', 'come', 'came', 'area', 'back', 'got', 'even', '``'])
```

```python
word_tokens = word_tokenize(b1_text_good) 
filtered_sentence = [w for w in word_tokens if not w in stop_words] 
  
cnt = Counter(filtered_sentence)
cnt.most_common(30)
```

```python
word_tokens = word_tokenize(b1_text_moderate) 
filtered_sentence = [w for w in word_tokens if not w in stop_words] 
  
cnt = Counter(filtered_sentence)
cnt.most_common(30)
```

```python
word_tokens = word_tokenize(b1_text_bad) 
filtered_sentence = [w for w in word_tokens if not w in stop_words] 
  
cnt = Counter(filtered_sentence)
cnt.most_common(50)
```
